```python
import pandas as pd
```

```python
pasta = 'E:/Documentos/bases/indice_conectar/população/'
```

```python
pop_2017 = pd.read_excel(pasta + 'POP2017_20210331.xls', sheet_name='base', dtype={'COD. UF': str, 'COD. MUNIC': str, 
                                                                                   'POPULAÇÃO ESTIMADA': int})
pop_2017['COD'] = pop_2017['COD. UF'] + pop_2017['COD. MUNIC']
pop_base = pop_2017.rename(columns={'POPULAÇÃO ESTIMADA' : 'pop_2017'})
```

```python
pop_2018 = pd.read_excel(pasta + 'POP2018_20210331.xls', sheet_name='base', dtype={'COD. UF': str, 'COD. MUNIC': str})
pop_2018['COD'] = pop_2018['COD. UF'] + pop_2018['COD. MUNIC']
pop_2018 = pop_2018.rename(columns={'POPULAÇÃO ESTIMADA' : 'pop_2018'})
pop_base = pd.merge(pop_base, pop_2018[['COD', 'pop_2018']], on='COD')
```

```python
pop_2019 = pd.read_excel(pasta + 'POP2019_20210331.xls', sheet_name='base', dtype={'COD. UF': str, 'COD. MUNIC': str})
pop_2019['COD'] = pop_2019['COD. UF'] + pop_2019['COD. MUNIC']
pop_2019 = pop_2019.rename(columns={'POPULAÇÃO ESTIMADA' : 'pop_2019'})
pop_base = pd.merge(pop_base, pop_2019[['COD', 'pop_2019']], on='COD')
```

```python
pop_2020 = pd.read_excel(pasta + 'POP2020_20210331.xls', sheet_name='base', dtype={'COD. UF': str, 'COD. MUNIC': str})
pop_2020['COD'] = pop_2020['COD. UF'] + pop_2020['COD. MUNIC']
pop_2020 = pop_2020.rename(columns={'POPULAÇÃO ESTIMADA' : 'pop_2020'})
pop_base = pd.merge(pop_base, pop_2020[['COD', 'pop_2020']], on='COD')
```

```python
pop_base.sort_values('COD', inplace=True)
```

```python
#Criando o código de 6 dígitos
pop_base['COD 6'] = pop_base['COD'].apply(lambda x: x[0:6])
```

```python
pop_base = pop_base[['UF', 'COD. UF', 'COD. MUNIC', 'COD', 'COD 6', 'NOME DO MUNICÍPIO',
                     'pop_2017', 'pop_2018', 'pop_2019', 'pop_2020']]
```

```python
# Removendo Brasília e Fernando de Noronha
pop_base_aux = pop_base[~pop_base['COD'].isin(['5300108', '2605459'])]                   
```

```python
#Incluir a categoria na base
categoria = pd.read_excel(r'E:\Documentos\bases\populacao\estimativa_dou_2020.xlsx', sheet_name='municipios_base', 
                          dtype={'munic_cod7':str})

pop_base_aux = pd.merge(pop_base_aux, categoria[['munic_cod7', 'categoria']], left_on='COD', right_on='munic_cod7', how='left')
```
